# NPV MINI PROJECT


**Problem Statement:** A Company is having its offices across all locations of India. These locations are categorized with respect to zones. The company has decided to give medical expenditure benefits to their employees. The company has collected the medical expenditure of all the employees for the year 2019. The company also collected some other health and family related data of all employees like number of dependents, age, bmi, smoking and alcoholic habits.

```python
# Reading the Data 
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns 

df_med_exp = pd.read_csv('emp_health.csv') # raeding a data from csv file 
df_med_exp.head() # first 5 rows
```

**1. How many employees above 50 age smoke, what is their average expenditure on health management?**

```python
df_med_exp[(df_med_exp['smoke']=='yes') & (df_med_exp['age']>50)]['expenditure'].mean()
```

**2. How many female employees did not consume alcohol daily. What is their proportion with respect to total female employees?**

```python
female_emp = ((df_med_exp['sex']=='female') & (df_med_exp['alcohol']!='daily')).sum()
female_emp
```

```python
total_female = df_med_exp[df_med_exp['sex']=='female']['sex'].count()
proportion = female_emp*100 / total_female
round(proportion,2).astype(str) + '%'
```

 **3. Display zone wise average expenditure of employees on health**

```python
avg_exp = df_med_exp.groupby('zone')['expenditure'].mean()
avg_exp
```

 **4. Employees of which zone did maximum expenditure on health**

```python
exp = df_med_exp.groupby(by='zone',sort=True)['expenditure'].sum()
exp.head(1)
```

 **5. What is the minimum expenditure done by any employee in organization**

```python
df_med_exp['expenditure'].min()
```

 **6. What is average expenditure of employees who have 2 or lesser dependents**

```python
round(df_med_exp[df_med_exp['dependent'] <= 2]['expenditure'].mean(),2)
```

 **7. How many employees are below age of 20 years. What is their average BMI**

```python
round(df_med_exp[df_med_exp['age'] < 20]['bmi'].mean(),2)
```

 **8. Create a dataframe "non_smoker" for all non-smoker male employees**

```python
non_smoker = df_med_exp[(df_med_exp['sex']=='male') & (df_med_exp['smoke']=='no')]
non_smoker
```

 **9. Create a dataframe "non_alcoholic" for all male employees who do not consume alcohol**

```python
non_alcoholic = df_med_exp[(df_med_exp['sex']=='male') & (df_med_exp['alcohol']=='no')]
non_alcoholic
```

 **10. Create a dataframe called "non_smoking_alcoholic" with the men employees who do not smoke and consume alcohol. The dataframe should contain columns id, smoke, alcohol**

```python
non_smoking_alcoholic = non_smoker.join(non_alcoholic.set_index(['id','smoke','alcohol']),on=['id','smoke','alcohol'], how="inner", lsuffix="_x", rsuffix="_y").loc[:,['id','smoke','alcohol']]
non_smoking_alcoholic
```

 **11. What is the count for each zone with respect to number of dependent**

```python
emp_health_zonewise = df_med_exp.groupby(['zone','dependent'])['id'].count()
emp_health_zonewise = emp_health_zonewise.reset_index()
emp_health_zonewise = emp_health_zonewise.rename(columns={'id':'count'})
emp_health_zonewise
```

**12. Classify the employees into 3 categories.
 Category 1 : employees below 18.5 bmi - underweight
 Category 2 : employees with bmi in the range of 18.5 to 24.9 - fit
 Category 3 : employee with bmi over 24.9 - overweight
 What is the average expenditure of employees in each category.
 Display the bar chart for the same**


```python
emp_underweight = df_med_exp[df_med_exp['bmi'] < 18.5]
emp_underweight['id'].count()
```

```python
emp_underweight_exp = emp_underweight['expenditure'].mean()
emp_underweight_exp
```

There are 20 underweight employees whoes average expenditue is Rs. 8852.20

```python
emp_fit = df_med_exp[(df_med_exp['bmi']>=18.5) & (df_med_exp['bmi']<=24.9)]
emp_fit['id'].count()
```

```python
emp_fit_exp = emp_fit['expenditure'].mean()
emp_fit_exp
```

There are 222 fit employees whoes average expenditue is Rs. 10379.49

```python
emp_overweight = df_med_exp[df_med_exp['bmi']>24.9]
emp_overweight['id'].count()
```

```python
emp_overweight_exp = emp_overweight['expenditure'].mean()
emp_overweight_exp
```

There are 1096 overweight employees whoes average expenditue is Rs. 13936.61

```python
bmi_type =['data_Underweight','data_Fit','data_Overweight']
exp_list=[emp_underweight_exp,emp_fit_exp,emp_overweight_exp]
df_chart_exp_bmi = pd.DataFrame(list(zip(bmi_type,exp_list)),columns=['BMI_Category','Average_Expenditure'])
df_chart_exp_bmi
```

```python
sns.barplot(x='BMI_Category',y='Average_Expenditure',data=df_chart_exp_bmi);
```
